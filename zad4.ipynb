```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hypersphere_volume(dimension: int, radius: float) -> float:
    return np.pi ** (dimension / 2) / math.gamma((dimension / 2) + 1) * radius ** dimension

def recursive_hypersphere_volume(dimension: int, radius: float) -> float:
    if dimension == 0:
        return 1
    elif dimension == 1:
        return 2 * radius
    elif dimension == 2:
        return np.pi * radius ** 2
    else:
        return (2 * np.pi * radius ** 2 / dimension) * recursive_hypersphere_volume(dimension - 2, radius)
```

```python
RADIUSES = [0.5, 1, 2]
DIMENSIONS = range(1, 51)

for radius in RADIUSES:
    volumes = [hypersphere_volume(d, radius) for d in DIMENSIONS]
    # volumes = [recursive_hypersphere_volume(d, radius) for d in DIMENSIONS]
    plt.plot(DIMENSIONS, volumes, label=f'Radius = {radius}')

    plt.xlabel('Dimension')
    plt.ylabel('Volume of Hypersphere')
    plt.title('Volume of Hyperspheres in Various Dimensions')
    plt.yscale("log")
    plt.xscale("log")
    plt.legend()
    plt.grid(True)
    plt.show()
```
